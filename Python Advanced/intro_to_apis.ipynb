### Problem 2:

I want to build the best Pokemon team I can and I need your help to do so! Using the API, build me a team of 6 pokemon who's base-stats are above 60. Return any additional information you think is needed to help understand why you chose that pokemon for the team!

```python
import requests

response = requests.get("https://pokeapi.co/api/v2/pokemon?limit=898")
pokemon_data = response.json()["results"]

team = [pokemon for pokemon in pokemon_data if sum(stat["base_stat"] for stat in requests.get(pokemon["url"]).json()["stats"]) > 360][:6]

for pokemon in team:
    print(f"Name: {pokemon['name']}")
    print(f"Base Stats: {requests.get(pokemon['url']).json()['stats']}")
    print(f"Abilities: {[ability['ability']['name'] for ability in requests.get(pokemon['url']).json()['abilities']]}\n")
```

### Problem 3:

Create a function that does the following:
* Input: pokemon_name(user_defined) -> str
* Output: list object containing a dictionary of values from the response object.
* Please include the type, the most current raritey level, at least one held item, and all stats

```python
import requests

def get_pokemon_info(pokemon_name):
    response = requests.get(f"https://pokeapi.co/api/v2/pokemon/{pokemon_name}")
    if response.status_code == 200:
        pokemon_data = response.json()
        return [{
            "type": [t["type"]["name"] for t in pokemon_data["types"]],
            "rarity": pokemon_data["sprites"]["version"]["name"],
            "held_item": [i["item"]["name"] for i in pokemon_data["held_items"]],
            "stats": {s["stat"]["name"]: s["base_stat"] for s in pokemon_data["stats"]}
        }]
    else:
        return []
```
