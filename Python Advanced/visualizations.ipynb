- What is the distribution of the `category_name` column?

```python
df['category_name'].value_counts().plot(kind='bar')
plt.title('Distribution of Category Names')
plt.xlabel('Category Name')
plt.ylabel('Count')

plt.xticks(fontsize=5)
plt.yticks(fontsize=7)
plt.show()

```

- Which country has the most projects?

```python
country_counts = df.groupby("country").size()

plt.bar(country_counts.index, country_counts.values)

# Add labels and title
plt.xlabel("Country")
plt.ylabel("Number of Projects")
plt.title("Number of Projects by Country")

plt.show()
```

- What are the top 20 Kickstarter projects?

- Which parent category has the most funded projects?

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the data into a pandas DataFrame
data = pd.read_csv(r"C:/Users/narim/Downloads/most_funded_feb_2023.csv")

# Group the data by parent category and calculate the total funding amount for each category
category_totals = data.groupby('parent_category')['funding_amount'].sum().reset_index()

# Sort the data by total funding amount in descending order
sorted_totals = category_totals.sort_values('funding_amount', ascending=False)

# Create a bar chart of total funding amount by parent category
fig, ax = plt.subplots()
ax.bar(sorted_totals['parent_category'], sorted_totals['funding_amount'])

# Add labels and title
ax.set_xlabel("Parent Category")
ax.set_ylabel("Total Funding Amount")
ax.set_title("Total Funding Amount by Parent Category")

# Rotate the x-axis labels for better readability
plt.xticks(rotation=90)

# Show the plot
plt.show()

# Group the data by parent category and calculate the total funding amount for each category
category_totals = data.groupby('parent_category')['funding_amount'].sum().reset_index()

# Sort the data by total funding amount in descending order
sorted_totals = category_totals.sort_values('funding_amount', ascending=False)

# Create a bar chart of total funding amount by parent category
fig, ax = plt.subplots()
ax.bar(sorted_totals['parent_category'], sorted_totals['funding_amount'])

# Add labels and title
ax.set_xlabel("Parent Category")
ax.set_ylabel("Total Funding Amount")
ax.set_title("Total Funding Amount by Parent Category")

# Rotate the x-axis labels for better readability
plt.xticks(rotation=90)

# Show the plot
plt.show()
```
